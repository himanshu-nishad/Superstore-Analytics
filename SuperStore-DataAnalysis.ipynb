# Data Analysis of Super-Store Dataset

In this project , Our Goal is to derive insights from the given data-set and Converting this data into more Accurate and informative

In this , We are going to Business Scenarios Related Question ,which is going build our concept
& Help in solving business related problem.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\91744\Downloads\Superstore-Dataset.csv")
```

```python
data.head(10)
```

```python
data = data.set_index('Row ID')
```

```python
data
```

```python
data.info()
```

```python
# we have changed  datatype of Order_date and ship_date because in was incorrect

data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True, errors='coerce')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], dayfirst=True, errors='coerce')
```

```python
data.info()
```

# Basic Business Scenarios Questions :
OUR GOAL IS TO ANALYSE THE DATA WITH BUSINESS-ACCUMEN

```python
# Q.1  What is the total Sales Generated in the Dataset?

total_sales = data['Sales'].sum()
print("Total Sales:", total_sales)


# It helped us understand the overall Revenue
```

```python
# Q.2 What is the Average Discount Offered??

avg_discount = data['Discount'].mean()
print("Average Discount:", avg_discount)


# It is showing 15% of average discount
```

```python
# Q.3 How many unique products are Sold???

unique_products = data['Product Name'].nunique()
print("Unique Products Sold:", unique_products)

# It helped us to understand Product Diversity
```

```python
# Q.4 Which Customer segment generated the Highest-Profit???

profit_by_segment = data.groupby('Segment')['Profit'].sum()
print(profit_by_segment.idxmax(), "is the most profitable segment.")

```

```python
# Q.5 What is the most freqently used shipping mode??

most_used_ship_mode = data['Ship Mode'].mode()[0]
print("Most Frequently Used Ship Mode:", most_used_ship_mode)
```

# Intermediate Business Scenario Questions:

```python
# Q.6 Which region has the highest number of orders?

orders_by_region = data['Region'].value_counts()
print("Region with Highest Orders:", orders_by_region.idxmax())

```

```python
# Q.7 What is the correlation between sales and profit?

correlation = data['Sales'].corr(data['Profit'])
print("Correlation between Sales and Profit:", correlation)
```

```python
# Q.8 Which city generates the highest revenue?

revenue_by_city = data.groupby('City')['Sales'].sum().sort_values(ascending=False)
print("City with Highest Revenue:", revenue_by_city.idxmax())

```

```python
# Q.9 What are the top 2 profitable product categories?

profitable_categories = data.groupby('Category')['Profit'].sum().sort_values(ascending=False).head(2)
print("Top Profitable Categories:\n", profitable_categories)
```

```python
# Q.10 How does sales distribution vary across months?

data['Order Month'] = data['Order Date'].dt.month
monthly_sales = data.groupby('Order Month')['Sales'].sum()
monthly_sales.plot(kind='bar', title='Monthly Sales Distribution')
plt.show()
```

# Advance Business Scenario Questions:

```python
# Q.11 What is the impact of Discount of Sales?
sns.scatterplot(x='Discount', y='Sales', data=data)
plt.title('Discount vs Sales')
plt.show()
```

```python
# Q.12 Predict which segment is likely to generate maximum profit in the next quarter?

quarterly_profit = data.resample('Q', on='Order Date')['Profit'].sum()
print("Segment likely to generate maximum profit:", quarterly_profit.idxmax())
```

```python
# Q.13 Which products consistently perform poorly?

low_profit_products = data.groupby('Product Name')['Profit'].mean().sort_values().head(10)
print("Poor Performing Products:\n", low_profit_products)
```

```python
# Q.14 How does profit vary across shipping modes?

profit_by_shipping = data.groupby('Ship Mode')['Profit'].sum()
profit_by_shipping.plot(kind='bar', title='Profit by Shipping Mode')
plt.show()
```

```python
# Q.15 Which combination of category and sub-category is most profitable?

cat_subcat_profit = data.groupby(['Category', 'Sub-Category'])['Profit'].sum().sort_values(ascending=False).head(1)
print("Most Profitable Combination:\n", cat_subcat_profit)
```

```python
# Q.16 What is the return on investment (ROI) for each region?

data['ROI'] = data['Profit'] / data['Sales']
roi_by_region = data.groupby('Region')['ROI'].mean()
print("ROI by Region:\n", roi_by_region)
```

```python
print(data.columns)
```

```python
data.columns = data.columns.str.strip()
```

```python
# Q.17 Perform cohort analysis on customer retention.

cohort = data.groupby(['Customer ID', data['Order Date'].dt.year])['Sales'].sum()
print("Customer Retention Cohort:\n", cohort.head())

```

```python
# Q.18 Which product categories have the highest variance in sales?

sales_variance = data.groupby('Category')['Sales'].var()
print("Variance in Sales by Category:\n", sales_variance)
```

# Analysis with Visualization & Insights

```python
numeric_dataset = data.select_dtypes(include='number')
```

```python
# Q.19 Create a heatmap showing correlation among numeric variables.

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_dataset.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Q.20 Analyze sales contribution by product categories as a pie chart.

data.groupby('Category')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', title='Sales by Category')
plt.ylabel('')
plt.show()
```

```python
# Q.21 Identify regions with consistent growth over years.

yearly_sales = data.groupby([data['Order Date'].dt.year, 'Region'])['Sales'].sum()
print(yearly_sales)
```

```python
# Q.22 Total Sales by Region

sales_by_region = data.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=sales_by_region, x='Region', y='Sales', palette='Blues_d')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q.23  Profit Distribution Across Categories

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Category', y='Profit', palette='Set2')
plt.title('Profit Distribution Across Categories')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#Detect variability and outliers in profit for each category.
```

```python
# Q.24 Discount vs. Profit Scatterplot

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Discount', y='Profit', alpha=0.6, hue='Category', palette='viridis')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()


# Analyze the impact of discount rates on profitability for each category.
```

```python
# Q.25 Monthly Sales Trend

data['Order Month'] = data['Order Date'].dt.to_period('M')  # Convert to period
data['Order Month'] = data['Order Month'].astype(str)       # Convert to string

monthly_sales = data.groupby('Order Month')['Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Order Month', y='Sales', marker='o', color='orange')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```

```python
# Q.26 Top 10 Profitable Products

top_products = data.groupby('Product Name')['Profit'].sum().nlargest(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, y='Product Name', x='Profit', palette='coolwarm')
plt.title('Top 10 Profitable Products')
plt.xlabel('Profit')
plt.ylabel('Product Name')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q.27  Profit by Segment and Region

segment_region_profit = data.groupby(['Segment', 'Region'])['Profit'].sum().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=segment_region_profit, x='Region', y='Profit', hue='Segment', palette='muted')
plt.title('Profit by Segment and Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.legend(title='Segment')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q.28 Sales Contribution by Categories

category_sales = data.groupby('Category')['Sales'].sum().reset_index()
plt.figure(figsize=(8, 6))
plt.pie(category_sales['Sales'], labels=category_sales['Category'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'salmon', 'lightgreen'])
plt.title('Sales Contribution by Category')
plt.show()
```

```python
# Q.29 Shipping Mode Preference

shipping_mode_count = data['Ship Mode'].value_counts().reset_index()
shipping_mode_count.columns = ['Ship Mode', 'Count']
plt.figure(figsize=(8, 6))
sns.barplot(data=shipping_mode_count, x='Ship Mode', y='Count', palette='pastel')
plt.title('Shipping Mode Preference')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q.30 Top Cities by Revenue

top_cities = data.groupby('City')['Sales'].sum().nlargest(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=top_cities, y='City', x='Sales', palette='plasma')
plt.title('Top 10 Cities by Revenue')
plt.xlabel('Sales')
plt.ylabel('City')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q.31  Categorical Breakdown of Sales and Profit

category_sales_profit = data.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
category_sales_profit_melted = category_sales_profit.melt(id_vars='Category', var_name='Metric', value_name='Value')
plt.figure(figsize=(10, 6))
sns.barplot(data=category_sales_profit_melted, x='Category', y='Value', hue='Metric', palette='Set1')
plt.title('Sales and Profit by Category')
plt.ylabel('Amount')
plt.xlabel('Category')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Metric')
plt.show()
```
